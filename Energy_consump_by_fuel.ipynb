```python
# imports library 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
file_path ='../World_energy_consumption/Raw_data_files/Energy_consumption_by_fuel.csv'
df_1= pd.read_csv(file_path, skiprows=2, encoding='unicode_escape')
df_1.head()
```

```python
rename_dict = {'Exajoules':'Country','Oil':'Oil 2021','Natural Gas':'Natural Gas 2021', 'Coal':'Coal 2021',
              'Nuclear energy': 'Nuclear energy 2021', 'Hydro electric': 'Hydro electric 2021', 'Renew- ables':'Renew- ables 2021', 'Total':'Total 2021',
             'Oil.1':'Oil 2022','Natural Gas.1':'Natural Gas 2022', 'Coal.1':'Coal 2022',
              'Nuclear energy.1': 'Nuclear energy 2022', 'Hydro electric.1': 'Hydro electric 2022', 'Renew- ables.1':'Renew- ables 2022', 'Total.1':'Total 2022' }
```

```python
df_1.rename(columns=rename_dict, inplace=True)
df = df_1.rename(columns=rename_dict)
df.head()
```

```python
df=df.drop([0,0])
```

```python
df.drop(["Unnamed: 15","Unnamed: 16", "Unnamed: 17", "Unnamed: 18"], axis=1, inplace=True)
df.head(10)
```

```python
df.set_index('Country', inplace=True)
df.head()
```

```python
null_rows= df.isna().any(axis=1)
```

```python
index_with_nulls = df.index[null_rows].tolist()

print(index_with_nulls)
```

```python
df_new = df.drop(index_with_nulls)
df_new.head(30)
```

```python
# Create the output file (CSV)

output_data_file = "../World_energy_consumption/Clean_data_files/Energy_cons_by_fuel.csv"


# Export the DataFrame to a CSV file
df_new.to_csv(output_data_file, index="Country")  
```
