```python
# imports library 

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

# Coal consumtion in Exajoules

```python
file_path ='../World_energy_consumption/Raw_data_files/Coal_consumtion_Exajoules.csv'
df_1= pd.read_csv(file_path, skiprows=2, encoding='unicode_escape')
df_1.head()
```

```python
rename_dict = {'Exajoules':'Country','2022.1':'Growth rate 2022-2021 in %','2012-22':'Growth rate 2022-2012 in %', '2022.2':'Share 2022 in %' }
```

```python
df_1.rename(columns=rename_dict, inplace=True)
df = df_1.rename(columns=rename_dict)
df.head()
```

```python
df=df.drop([0,0])
```

```python
df.shape
```

```python
df.count()
```

```python
df.isnull().sum()
```

```python
df.set_index('Country', inplace=True)
```

```python
df.head()
```

```python
null_rows= df.isna().any(axis=1)
```

```python
index_with_nulls = df.index[null_rows].tolist()

print(index_with_nulls)
```

```python
df_new = df.drop(index_with_nulls)
df_new.head()
```

```python
df_new.isnull().sum()
```

```python
df_new.dtypes
```

```python
columns_to_convert = df_new.columns[:-3]
```

```python
for column in columns_to_convert:
    df_new[column] = pd.to_numeric(df_new[column], errors='coerce')
    
df_new.dtypes
```

```python
# Function to convert percentage strings to float and handle non-numeric values
def convert_percent_to_float(df_new_column):
    # Replace '%' and handle non-numeric entries
    return pd.to_numeric(df_new_column.str.replace('%', '').replace('-', np.nan), errors='coerce')

# Applying the conversion function to the last three columns
df_new['Growth rate 2022-2021 in %'] = convert_percent_to_float(df_new['Growth rate 2022-2021 in %'])
df_new['Growth rate 2022-2012 in %'] = convert_percent_to_float(df_new['Growth rate 2022-2012 in %'])
df_new['Share 2022 in %'] = convert_percent_to_float(df_new['Share 2022 in %'])

# Output the updated Data type in the dataframe
df_new.dtypes
```

```python
# Dropping rows where any cell has NaN
df_new.fillna(0, inplace=True)

# Output the updated DataFrame
df_new.head(30)
```

```python
# Create the output file (CSV)

output_data_file = "../World_energy_consumption/Clean_data_files/Coal_consumption_Exajoules.csv"


# Export the DataFrame to a CSV file
df_new.to_csv(output_data_file, index="Country")  
```

```python
# Now, transforming the data to store in postgres
file_path ='../World_energy_consumption/Clean_data_files/Coal_consumption_Exajoules.csv'
df_new= pd.read_csv(file_path, skiprows=0, encoding='unicode_escape')
df_new.head()
```

```python
# dropping the last three rows as these are not important for the analysis
columns_to_drop = df.columns[-3:]
df_new = df_new.drop(columns=columns_to_drop, axis=1)
df_new.head()
```

```python
df_new = df_new.melt(id_vars=['Country'], var_name='year', value_name='energy_consumption')
```

```python
# Create the output file (CSV)

output_data_file = "../World_energy_consumption/Transformed_data_files/Coal_consumption_Exajoules.csv"


# Export the DataFrame to a CSV file
df_new.to_csv(output_data_file, index="Country")  
```
